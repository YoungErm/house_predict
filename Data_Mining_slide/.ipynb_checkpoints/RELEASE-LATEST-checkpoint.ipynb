#### 2.1 RegionÂú∞Âå∫ÂàÜÊûê

```python
df_house_count = df.groupby('Region')['Price'].count().sort_values(ascending=False).to_frame().reset_index()
df_house_mean = df.groupby('Region')['PerPrice'].mean().sort_values(ascending=False).to_frame().reset_index()

f, [ax1, ax2, ax3] = plt.subplots(3, 1, figsize=(20, 15))
sns.barplot(x='Region', y='PerPrice', palette="Blues_d", data=df_house_mean, ax=ax1)
ax1.set_title('Âåó‰∫¨ÂêÑÂ§ßÂå∫ÊàøÂ±ãÊØèÂπ≥Á±≥Âçï‰ª∑ÂØπÊØî', fontsize=12)
ax1.set_xlabel('Âå∫Âüü', fontsize=12)
ax1.set_ylabel('ÊØèÂπ≥Á±≥Âçï‰ª∑', fontsize=12)

sns.barplot(x='Region', y='Price', palette="Greens_d", data=df_house_count, ax=ax2)
ax2.set_title('Âåó‰∫¨ÂêÑÂ§ßÂå∫ÊàøÂ±ãÊï∞ÈáèÂØπÊØî', fontsize=12)
ax2.set_xlabel('Âå∫Âüü', fontsize=12) 
ax2.set_ylabel('Êï∞Èáè', fontsize=12)

sns.boxplot(x='Region', y='Price', data=df, ax=ax3)
ax3.set_title('Âåó‰∫¨ÂêÑÂ§ßÂå∫ÊàøÂ±ãÊàøÂ±ãÊÄª‰ª∑', fontsize=12)
ax3.set_xlabel('Âå∫Âüü', fontsize=12)
ax3.set_ylabel('ÊàøÂ±ãÊÄª‰ª∑', fontsize=12)

# plt.show()
plt.savefig("Region.png")
```

    ‰æãÂ¶ÇÔºöÊüê‰∏™ÂõõÂêàÈô¢ÁöÑÊúùÂêëÊòØ:`‰∏úÂçóÂåó`ÔºåÈÇ£‰πàÂØπÂ∫îÁöÑÂêëÈáè‰∏∫$[1,1,0,1]$|

```python
for i in range(len(data["Direction"])):
    dire = ["‰∏ú","Âçó","Ë•ø","Âåó"]
    
    for j in dire:
        if j in data.at[i,"Direction"]:
            data.at[i,j] = 1
        else:
            data.at[i,j] = 0

data.drop(['Direction'], axis=1, inplace=True) 
            
data.to_csv("Direction_split.csv")
data.iloc[:5,29:]
```

```python
# ÁªüËÆ°Áº∫Â§±ÂÄº
def calc_mis_val(df):
    cols = df.columns
    mis_val = df.isnull().sum()
    mis_val_pct = round(100 * mis_val / df.shape[0], 2)
    mis_val_df = pd.DataFrame({'mis_val':mis_val, 'mis_val_pct(%)':mis_val_pct}, index=cols)
    mis_val_df = mis_val_df[mis_val_df['mis_val'] != 0].sort_values('mis_val', ascending=False)
    print('ÊÄªÂàóÊï∞Ôºö', df.shape[1])
    print('Âê´Áº∫Â§±ÂÄºÂàóÊï∞Ôºö', mis_val_df.shape[0])
    return mis_val_df

all_mis_val = calc_mis_val(data)
all_mis_val
```

```python
# Âª∫Á´ãÂü∫Á∫øÊ®°Âûã
lgb = LGBMRegressor(objective='regression', random_state=50)
xgb = XGBRegressor(objective='reg:squarederror', random_state=50)
ridge = make_pipeline(RobustScaler(), RidgeCV(cv=kf))
svr = make_pipeline(RobustScaler(), SVR())
gbr = GradientBoostingRegressor(random_state=50)
rfc = RandomForestRegressor(random_state=50)
# Âü∫Á∫øÊ®°ÂûãËØÑ‰º∞
models = [lgb, xgb, ridge, svr, gbr, rfc]
model_names = ['lgb','xgb','ridge','svr','gbr','rfc']
scores = {}
```

$y_i$ÊòØÂÆûÈôÖÁöÑËßÇÂØüÂÄºÔºå $\hat{y_i}$ÊòØÊ®°ÂûãÈ¢ÑÊµãÁöÑËæìÂá∫ÂÄº

Âπ≥ÂùáËßÇÂØüÂÄº:
$$\overline{y} = \frac{1}{n}\sum^n_{i=1}y_i$$

ÊÄª‰ΩìÁ¶ªÂ∑ÆÂπ≥ÊñπÂíå:
$$SS_{total} = \sum^n_{i=1}(y_i-\overline{y})^2$$

ÊÆãÂ∑ÆÂπ≥ÊñπÂíå:
$$SS_{residual} = \sum^n_{i=1}(y_i-\hat{y_i})^2$$



**Ë°°ÈáèÊåáÊ†áÔºö**

**RMSE:**(ÂùáÊñπÊ†πËØØÂ∑Æ) 
    $$RMSE = \sqrt{\frac{1}{n}\sum^n_{i=1}(y_i - \hat{y_i})^2}$$
    
**RMSE**ÂÆπÊòìÂèóÂà∞Âõ†ÂèòÈáèÔºàÁõÆÊ†áÔºâÂíåËá™ÂèòÈáèÔºàÁâπÂæÅÔºâÈáèÁ∫≤Â§ßÂ∞èÁöÑÂΩ±Âìç:
- ÊØîÂ¶ÇËØ¥ÔºåAÊ®°Âûã‰∏≠ÁöÑÂõ†ÂèòÈáèÁöÑÊï∞ÈáèÁ∫ßÈÉΩÊòØËØ∏Â¶Ç1000„ÄÅ2000‚Ä¶ËøôÊ†∑ÁöÑÊï∞ÂÄº
-     ËÄåBÊ®°Âûã‰∏≠Âõ†ÂèòÈáèÈÉΩÊòØ1„ÄÅ2‚Ä¶ËøôÊ†∑Êï∞ÈáèÁ∫ßÁöÑÂÄº„ÄÇ
- ÂæàÊòæÁÑ∂ÔºåÁõ∏ÂØπ‰∫éÊ®°ÂûãBÔºå**AÊ®°Âûã**‰∏≠**RMSE**ÂæàÂèØËÉΩ‰ºö**ÊØîËæÉÂ§ß**„ÄÇ

‰ΩÜËøô‰∏çËÉΩËØ¥ÊòéAÊ®°ÂûãÁöÑÊ≥õÂåñËÉΩÂäõÂ∞±ÊØîBÊ®°ÂûãÂ∑Æ„ÄÇ

ÊâÄ‰ª•**R2 Score**Â∞ÜÂõûÂΩíÁªìÊûúÂΩíÁ∫¶Âà∞‰∫Ü0~1Èó¥

**R2 Score:** 
    $$R^2 Score = 1- \frac{SS_{residual}}{SS_{total}}$$

```python
# ÂÆö‰πâË°°ÈáèÊåáÊ†áÂáΩÊï∞
def cv_rmse(model, X=X):
    rmse = np.sqrt(-cross_val_score(model, X, Y, scoring='neg_mean_squared_error', cv=kf,n_jobs=-1))
    return rmse

def r2_score(model):
    model.fit(Xtrain, Ytrain)
    score = model.score(Xtest, Ytest)
    return score
```

**ËØÑ‰º∞ÊåáÊ†áÔºö** $R^2 score$

```python
# ËøêË°åÈúÄË¶Å2min 36s
for i, model in enumerate(models):
    score = r2_score(model)
    print("{} r2_score: {:.4f}".format(model_names[i],score))
    scores[model_names[i]] = score

r2_df = pd.DataFrame(scores, index=['r2_score'])
r2_df.sort_values('r2_score', axis=1, inplace=True)
r2_df.to_csv("r2_score.csv")
r2_df
```

**ËØÑ‰º∞ÊåáÊ†áÔºö** $ÂùáÊñπËØØÂ∑Æ RMSE$

```python
# ËøêË°åÈúÄË¶ÅÔºö38min 45s
for i, model in enumerate(models):
    score = cv_rmse(model)
    print('{} rmse score: {:.4f}, rmse std: {:.4f}'.format(model_names[i], score.mean(), score.std()))
    scores[model_names[i]] = (score.mean(), score.std())
    
rmse_df = pd.DataFrame(scores, index=['rmse_score','rmse_std'])

rmse_df.sort_values('rmse_score', axis=1, inplace=True)
rmse_df
```

```python
sum_info = r2_df.append(rmse_df)
sum_info.to_csv('sum_info.csv')
sum_info
```
